```
# Import Requests module to fetch data from a website
import requests as rq
```

```
# Import bs4 for data extraction , re for regex and collections to calc word frequency
import bs4
import re
import collections as col
```

```
# store all data in res variable
res=rq.get("https://en.wikipedia.org/wiki/Machine_learning")
```

```
res.text
```

```
# create a soup object
soup=bs4.BeautifulSoup(res.text,'lxml')
```

```
sel=soup.select('body')
```

```
text=sel[0].getText()
```

```
# fetch all words from website data and store in a list
words=re.findall('\w+',text)
```

```
# Calculate word count
wc=len(words)
```

```
# Find word frequency
word_counts=col.Counter(words)
```

```
word_counts
```

```
# store words and their counts in seperate lists
w=[]
co=[]
for i,c in sorted(word_counts.items(),key=lambda x:x[-1],reverse=True):
  w.append(i)
  co.append(c)
```

```
# plot graphs
import matplotlib.pyplot as plt
```

```
# Plot Bar graph for top 20 repeated words.
plt.figure(figsize=(20,3))
plt.xlabel('Words')
plt.ylabel('Count')
plt.bar(w[0:20],co[0:20])
```

```
# Plotting pie chart for top 20 repeated words
plt.pie(co[0:20],labels=w[0:20],radius=1.8)
```

```
# Plotting scatter graph for words having presence more than 3 times
plt.xlabel('count')
plt.ylabel('words')
plt.scatter(co[0:492],w[0:492])
```
